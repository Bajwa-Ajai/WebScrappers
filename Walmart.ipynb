```python
import csv
import requests
import json
```

```python
def put_resp(resp):
  data=json.loads(resp)
  for item in data['items']:
    product={}
    try:
      product['link']='https://www.walmart.com'+item['productPageUrl']
    except:
      product['link']=''
    try:
      product['title']=item['title']
    except:
      product['title']=''
    try:
      product['sub-category']=item['productCategory']
    except:
      product['sub-category']=''
    try:
      product['image']=item['imageProps']['src']
    except:
      product['image']=''
    try:
      product['price']=item['primaryOffer']['offerPrice']
    except:
      product['price']=''
    try:
      product['description']=item['description'].replace('<li>',' ').replace('</li>',' ')
    except:
      product['description']=''  
    product['category']='Grooming'
    print(product)
    write_csv(product)
```

```python
def write_csv(data):
  with open('walmart5(Grooming).csv','a') as csvfile:
    writer=csv.writer(csvfile)
    row=[data['category'],data['sub-category'],data['link'],data['title'],data['description'],data['price'],data['image']]

    writer.writerow(row)
```
